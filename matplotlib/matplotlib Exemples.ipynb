```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
values = [1,2,3,4,5,6]
squares = [value ** 2 for value in values]
squares
```

```python
plt.plot(values, squares)
plt.show
```

```python
plt.scatter(values, squares)
plt.show
```

```python
values = list(range(100))
pows = [value ** 3 for value in values ]

plt.scatter(values, pows)
plt.show()
```

```python
plt.scatter(values, pows, c =pows, cmap=plt.cm.Blues, edgecolors='none', s=10)
```

```python
values = list(range(10))
squares = [value ** 2 for value in values ]
coubes = [value ** 3 for value in values ]

fig, (ax1,ax2) = plt.subplots(2, 1, sharey=True)
ax1.scatter(values, squares)
ax1.set_title("mops1")
ax2.plot(values, coubes)
ax2.set_title("mops2")

```

```python
data = {
    'Year': [2001, 2002, 2003, 2004, 2005, 2006],
    'Sales': [100, 101, 102, 103, 104, 105]

}
df = pd.DataFrame(data)
df.plot(x='Year', y='Sales', kind = 'line')
plt.show()
```

```python
data = {
    'Year': [2001, 2002, 2003, 2004, 2005, 2006],
    'Sales': [100, 101, 102, 103, 104, 105]

}
df = pd.DataFrame(data)
df.plot(y='Sales', kind = 'pie', labels=df['Year'])
plt.title('Yearly Sales')
plt.show()
```

```python
data = {
    'Age': [32, 32, 12, 14, 16, 17, 19, 9, 3, 9],

}
df = pd.DataFrame(data) 
df.plot(y='Age',kind = 'hist')
plt.show()
```
