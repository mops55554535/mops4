```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
values = [1,2,3,4,5,6]
squares = [value ** 2 for value in values]
squares
```

```python
plt.plot(values, squares)
plt.show
```

```python
plt.scatter(values, squares)
plt.show
```

```python
values = list(range(100))
pows = [value ** 3 for value in values ]

plt.scatter(values, pows)
plt.show()
```

```python
plt.scatter(values, pows, c =pows, cmap=plt.cm.Blues, edgecolors='none', s=10)
```

```python
values = list(range(10))
squares = [value ** 2 for value in values ]
coubes = [value ** 3 for value in values ]

fig, (ax1,ax2) = plt.subplots(2, 1, sharey=True)
ax1.scatter(values, squares)
ax1.set_title("mops1")
ax2.plot(values, coubes)
ax2.set_title("mops2")

```
